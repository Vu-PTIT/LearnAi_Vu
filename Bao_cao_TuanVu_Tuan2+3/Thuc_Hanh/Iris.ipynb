```python
df.columns =[col.replace(' (cm)', '').replace(' ','_') for col in df.columns]
df.head()
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
scaler= StandardScaler()
features = df.drop('target', axis=1)  
features_scaled = scaler.fit_transform(features)
df_scaled = pd.DataFrame(features_scaled, columns=features.columns)
df_scaled['target'] = df['target']
df_scaled
```

```python
df_scaled['sepal_area'] = df_scaled['sepal_length'] * df_scaled['sepal_width']
df_scaled
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df_scaled.corr(), annot=True, cmap='coolwarm')
plt.title("Ma trận tương quan của các feature")
```

```python
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split,cross_val_score
```

```python

X = df_scaled.drop(['target'], axis=1)
y = df_scaled['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python

dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, y_train)
y_pred_dt = dt_clf.predict(X_test)
dt_clf.score(X_test, y_test)
```

```python

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred_lr = log_reg.predict(X_test)
log_reg.score(X_test, y_test)
```

```python

X_reg = df_scaled.drop(['target', 'sepal_length'], axis=1)
y_reg = df_scaled['sepal_length']

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg,  random_state=0)

lin_reg = LinearRegression()
lin_reg.fit(X_train_reg, y_train_reg)
y_pred_reg = lin_reg.predict(X_test_reg)
lin_reg.score(X_test_reg, y_test_reg)
```

```python
from sklearn.metrics import  accuracy_score,recall_score,precision_score,f1_score
```

```python

print("Logistic Regression:")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Precision:", precision_score(y_test, y_pred_lr, average='weighted'))
print("Recall:", recall_score(y_test, y_pred_lr, average='weighted'))
print("F1 Score:", f1_score(y_test, y_pred_lr, average='weighted'))
print("\n")


print("Decision Tree Classifier:")
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Precision:", precision_score(y_test, y_pred_dt, average='weighted'))
print("Recall:", recall_score(y_test, y_pred_dt, average='weighted'))
print("F1 Score:", f1_score(y_test, y_pred_dt, average='weighted'))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print("Linear Regression (Hồi quy):")
print("Mean Absolute Error:", mean_absolute_error(y_test_reg, y_pred_reg))
print("Mean Squared Error:", mean_squared_error(y_test_reg, y_pred_reg))
print("R-squared:", r2_score(y_test_reg, y_pred_reg))
```
