```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic= sns.load_dataset('titanic')
titanic
```

```python
titanic['age'].fillna(titanic['age'].median(),inplace=True)
titanic[titanic['age']=='NaN']
```

```python
titanic.drop(columns=['deck', 'class', 'who', 'adult_male', 'embark_town', 'alive', 'alone'], inplace=True)
```

```python
titanic['embarked'].fillna(titanic['embarked'].mode()[0],inplace=True)
titanic['embarked']
```

```python
titanic
```

```python
titanic['family']=titanic['sibsp']+titanic['parch']+1
titanic
```

```python
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.tree import ExtraTreeClassifier
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
X_cls = titanic.drop(columns=['survived'])
y_cls = titanic['survived']
X_reg = titanic.drop(columns=['fare', 'survived'])
y_reg = titanic['fare']
```

```python
X_cls_train, X_cls_test, y_cls_train, y_cls_test = train_test_split(
    X_cls, y_cls, test_size=0.2, random_state=42
)
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)
```

```python
numeric_features = ['age', 'sibsp', 'parch', 'family']
numeric_transformer = Pipeline([('scaler', StandardScaler())])

categorical_features = ['pclass', 'sex', 'embarked']
categorical_transformer = Pipeline([('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

```python
cls_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])
cls_pipeline.fit(X_cls_train, y_cls_train)
cls_pipeline.score(X_cls_test, y_cls_test)
```

```python
reg_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
reg_pipeline.fit(X_reg_train, y_reg_train)
reg_pipeline.score(X_reg_test, y_reg_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
y_pred_cls = cls_pipeline.predict(X_cls_test)
print("Classification Metrics:")
print(f"Accuracy: {accuracy_score(y_cls_test, y_pred_cls):.4f}")
print(f"Precision: {precision_score(y_cls_test, y_pred_cls):.4f}")
print(f"Recall: {recall_score(y_cls_test, y_pred_cls):.4f}")
print(f"F1 Score: {f1_score(y_cls_test, y_pred_cls):.4f}")
print("\n", classification_report(y_cls_test, y_pred_cls))
```

```python
y_pred_reg = reg_pipeline.predict(X_reg_test)
print("Regression Metrics:")
print(f"MAE: {mean_absolute_error(y_reg_test, y_pred_reg):.4f}")
print(f"MSE: {mean_squared_error(y_reg_test, y_pred_reg):.4f}")
print(f"R2: {r2_score(y_reg_test, y_pred_reg):.4f}")
```

```python
cv_cls = cross_val_score(cls_pipeline, X_cls, y_cls, cv=5, scoring='accuracy')
print("Accuracy:", cv_cls.mean())

cv_reg = cross_val_score(reg_pipeline, X_reg, y_reg, cv=5, scoring='r2')
print("R2:", cv_reg.mean())
```
